```python
# Importing Librariesrr
import cv2
import mediapipe as mp
import pyautogui

# Used to convert protobuf message to a dictionary.
from google.protobuf.json_format import MessageToDict

# Initializing the Model
mpHands = mp.solutions.hands
hands = mpHands.Hands()
drawing_utils=mp.solutions.drawing_utils
screen_width, screen_height = pyautogui.size()
index_y=0
x1=x2=y1=y2=0

# Start capturing video from webcam
cap = cv2.VideoCapture(0)

while True:
    # Read video frame by frame
    success, img = cap.read()

    frame_height,frame_width,_=img.shape
    print(frame_width,frame_height)

    # Flip the image(frame)
    img = cv2.flip(img, 1)

    # Convert BGR image to RGB image
    imgRGB = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    # Process the RGB image
    results = hands.process(imgRGB)

    # If hands are present in image(frame)
    if results.multi_hand_landmarks:

        for hand in results.multi_hand_landmarks:
            drawing_utils.draw_landmarks(img,hand)
        if len(results.multi_handedness) == 2:
            # Display 'Both Hands' on the image
                        cv2.putText(img, 'Both Hands', (250, 50),
                        cv2.FONT_HERSHEY_COMPLEX,
                        0.9, (0, 255, 0), 2)

        # If any hand present
        else:
            for i in results.multi_handedness:

                # Return whether it is Right or Left Hand
                label = MessageToDict(i)['classification'][0]['label']
                for id, landmark in enumerate(hand.landmark):




                 if label == 'Left':




                     if id == 8:
                       x = int(landmark.x * frame_width)
                       y = int(landmark.y * frame_height)
                       cv2.circle(img=img, center=(x,y), radius=4, color=(0, 255, 250),thickness=3)
                       x1 = x
                       y1 = y

                     if id == 4:
                       x = int(landmark.x * frame_width)
                       y = int(landmark.y * frame_height)
                       cv2.circle(img=img, center=(x,y), radius=4, color=(0, 255, 250),thickness=3)
                       x2 = x
                       y2 = y

                       dist = ((x2 - x1) ** 2 + (y2 - y1) ** 2) ** 0.5 // 4

                       cv2.line(img, (x1, y1), (x2, y2), (0, 255, 255), 5)

                       if dist > 30:
                           pyautogui.press("volumeup")

                       else:
                           pyautogui.press("volumedown")

                       cv2.putText(img, label + ' Hand',(20, 50),cv2.FONT_HERSHEY_COMPLEX,0.9, (0, 255, 0), 2)

        if label == 'Right':

            for id, landmark in enumerate(hand.landmark):


                   if id == 8:
                    x = int(landmark.x * frame_width)
                    y = int(landmark.y * frame_height)
                    print(x,y)
                    cv2.circle(img=img, center=(x, y), radius=4, color=(0, 255, 250), thickness=3)
                    index_x = screen_width / frame_width * x
                    index_y = screen_height / frame_height * y
                    pyautogui.moveTo(index_x,index_y)

            for id, landmark in enumerate(hand.landmark):

                    if id == 4:
                        x = int(landmark.x * frame_width)
                        y = int(landmark.y * frame_height)
                        cv2.circle(img=img, center=(x, y), radius=10, color=(0, 255, 255))
                        thumb_x = screen_width / frame_width * x
                        thumb_y = screen_height / frame_height * y
                        print('outside', abs(index_y - thumb_y))
                        if abs(index_y - thumb_y) < 50:
                            pyautogui.click()
                            pyautogui.sleep(1)
                            pyautogui.moveTo(index_x, index_y)



                    cv2.putText(img, label + ' Hand', (460, 50),cv2.FONT_HERSHEY_COMPLEX,0.9, (0, 255, 0), 2)

    # Display Video and when 'q'
    # is entered, destroy the window
    cv2.imshow('Image', img)
    if cv2.waitKey(1) & 0xff == ord('q'):
        break
```
